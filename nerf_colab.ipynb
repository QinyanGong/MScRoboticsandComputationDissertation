```
%cd /content/nerf_pl

import os
# set training configurations here
os.environ['ROOT_DIR'] = "/content/drive/My Drive/nerf/dataset"
                         # directory containing the data
os.environ['IMG_W'] = "450" # image width (do not set too large)
os.environ['IMG_H'] = "360" # image height (do not set too large)
os.environ['NUM_EPOCHS'] = "10" # number of epochs to train (depending on how many images there are,
                                # 20~30 might be enough)
os.environ['EXP'] = "C3VD" # name of the experience (arbitrary)

!python train.py \
   --dataset_name llff \
   --root_dir "$ROOT_DIR" \
   --N_importance 64 --img_wh $IMG_W $IMG_H \
   --num_epochs $NUM_EPOCHS --batch_size 256 \
   --optimizer adam --lr 5e-4 \
   --lr_scheduler cosine \
   --exp_name $EXP
```
